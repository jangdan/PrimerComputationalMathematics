## Latitude
Latitude is the angular distance of a place north or south of the earth's equator, or of the equator of a celestial object, usually expressed in degrees, minutes and seconds.

1 degree latitude is about 111 km anywhere on earth.

Lines of latitude are numbered from 0째 at the equator to 90째N or +90 at the North Pole.
Lines of latitude are numbered from 0째 at the equator to 90째S or -90 at the South Pole.

```python
plt.figure()
plt.subplot(111, projection="aitoff")
plt.title("Latitude")
plt.grid(True)
```

## Longitude
The distance of 1 degree longitude varies from 111 km at equator to 0 at poles, half way approx. = 55.8 km at 60 degrees latitude. 

Longitude varies as cos罐, where 罐 is the latitude (e.g. long at 80째N = 111cos(80째) = 19km)

Lines of longitude begin at the Prime Meridian (Greenwich).

60째 W is the 60째 line of longitude west of the Prime Meridian. Is also written as -60째
60째 E is the 60째 line of longitude east of the Prime Meridian. Is also written as +60째

Sometimes you will see longitude on a scale 0째 to 360째. The scale is measured clockwise (so 10째W == 350째 == -10째)

```python
plt.figure()
plt.subplot(111, projection="aitoff")
plt.title("Longitude")
plt.grid(True)
```

```python
# Plot to covert longitude degree to kilometer

x = np.arange(0, 90, 0.1)
y = 111*np.cos(x*(np.pi)/180)

plt.plot(x, y)
plt.title("Coverting longitude degree to kilometer")

plt.show()
```
