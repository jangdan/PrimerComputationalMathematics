```python
###### CONVERGENCE OF SEQUENCES ######
###### Lecture 6, slide 11, 13, 14 ######
k = Symbol('k')
print("As k->infinity, the sequence {k} tends to: %f" % limit(k, k, oo) ) 
# The 'oo' here is SymPy's notation for infinity

print("As k->infinity, the sequence {1/k} tends to: %f" % limit(1.0/k, k, oo)) 

print("As k->infinity, the sequence {exp(1/k)} tends to: %f" % limit(exp(1.0/k), k, oo)) 

print("As k->infinity, the sequence {(k**3 + 2*k - 4)/(k**3 + 1)} tends to: %f" 
      % limit((k**3 + 2*k - 4)/(k**3 + 1), k, oo)) 
```

```python
###### SERIES ######
###### Lecture 6, slide 15 ######
# Using list comprehension:
print("The sum of 3*k + 1 (from k=0 to k=4) is: %f" % sum([3*k + 1 for k in range(0,5)])) 
# Note: we could also use the nsum function (part of the module mpmath): 
# import mpmath
# print mpmath.nsum(lambda k: 3*k + 1, [0, 4])

x = 1
print("The sum of (x**k)/(k!) from k=0 to k=4, with x = 1, is: %f" % sum([x**k/factorial(k) for k in range(1,5)])) 
```

```python
###### RATIO TEST ######
###### Lecture 6, slide 27 ######
# A divergent example
k = Symbol('k')
f = (2**k)/(3*k)
f1 = (2**(k+1))/(3*(k+1))
ratio = f1/f

lim = limit(ratio, k, oo) 
print("As k -> infinity, the ratio tends to: %f" % lim) 
if(lim < 1.0):
    print("The series converges") 
elif(lim > 1.0):
    print("The series diverges") 
else:
    print("The series either converges or diverges") 

# A converging example
f = (2**k)/(5**k)
f1 = (2**(k+1))/(5**(k+1))
ratio = f1/f
```

```python
###### POWER SERIES ######
###### Lecture 6, slide 30 ######
k = Symbol('k')
x = Symbol('x')

a = 1.0/k
f = a*(x**k)

a1 = 1.0/(k+1)
f1 = a1*(x**(k+1))

ratio = abs(a/a1)
R = limit(ratio, k, oo)
print("The radius of convergence (denoted R) is: %f" % R) 

x = 0.5
if(abs(x) < 1):
    print("The series converges for |x| = %f (< R)" % abs(x)) 
elif(abs(x) > 1):
    print("The series diverges for |x| = %f (> R)" % abs(x)) 
else:
    print("The series either converges or diverges for |x| = %f (== R)\n" % abs(x)) 
```

```python
###### USEFUL SERIES ######
###### Lecture 6, slide 34 ######
x = Symbol('x')
r = Symbol('r')

# Note: the optional argument 'n' allows us to truncate the series
# after a certain order of x has been reached.
print("1/(1+x) = ", series(1.0/(1.0+x), x, n=4)) 
print("1/(1-x) = ", series(1.0/(1.0-x), x, n=4)) 
print("ln(1+x) = ", series(ln(1.0+x), x, n=4)) 
print("exp(x) = ", series(exp(x), x, n=4)) 
print("cos(x) = ", series(cos(x), x, n=7)) 
print("sin(x) = ", series(sin(x), x, n=8)) 
```

```python
### plot taylor series for ln(1+x) with different number of terms n

def ln_taylor(x, n):
    y = 0
    for i in range(1,n):
        y += (-1)**(i+1) * (x**i)/(factorial(i))
    return y

x = numpy.array([0.01*i for i in range(-50, 50)])

ln = numpy.log(1+x)

n = numpy.array([i for i in range(1, 6)])

y = {}

for i in range(len(n)):
    y['n='+str(n[i])] = []
    for j in range(len(x)):
        y['n='+str(n[i])].append(ln_taylor(x[j], n[i]))

colour = ['c', 'b', 'g', 'y', 'r']        

# set up figure
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)
ax1.margins(0.1)

# plot figure
for i in range(len(n)):
    ax1.plot(x, y['n='+str(n[i])], colour[i], label='n=%.d' % (n[i]))
ax1.plot(x, ln, 'k', label='y=exp(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('A plot to compare the function ln(1+x) with its Taylor series', fontsize=16)
ax1.legend(loc='best', fontsize=14)
ax1.grid(True)
```

```python
### plot taylor series for exp(x) with different number of terms n

def exp_taylor(x, n):
    y = 0
    for i in range(n):
        y += x**i / factorial(i)
    return y


x = numpy.array([0.01*i for i in range(-500, 500)])

expx = numpy.exp(x)

n = numpy.array([i for i in range(1, 6)])

y = {}

for i in range(len(n)):
    y['n='+str(n[i])] = []
    for j in range(len(x)):
        y['n='+str(n[i])].append(exp_taylor(x[j], n[i]))

colour = ['c', 'g', 'b', 'y', 'r']        

# set up figure
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)
ax1.margins(0.1)

# plot figure
for i in range(len(n)):
    ax1.plot(x, y['n='+str(n[i])], colour[i], label='n=%.d' % (n[i]))
ax1.plot(x, expx, 'k', label='y=exp(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('A plot to compare exp(x) with its Taylor series', fontsize=16)
ax1.legend(loc='best', fontsize=14)
ax1.grid(True)
```

```python
### plot taylor series for sin(x) with different number of terms n

def sin_taylor(x, n):
    y = 0
    for i in range(n):
        y += (-1)**(i)*(x**(2*i+1) / factorial(2*i+1))
    return y

x = numpy.array([0.01*i for i in range(-600, 600)])

sinx = numpy.sin(x)

n = numpy.array([i for i in range(1, 6)])

y = {}

for i in range(len(n)):
    y['n='+str(n[i])] = []
    for j in range(len(x)):
        y['n='+str(n[i])].append(sin_taylor(x[j], n[i]))

colour = ['r', 'y', 'g', 'b', 'c']        

# set up figure
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)
ax1.margins(0.1)

# plot figure
for i in range(len(n)):
    ax1.plot(x, y['n='+str(n[i])], colour[i], label='n=%.d' % (n[i]))
ax1.plot(x, sinx, 'k', label='y=sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('A plot to compare sin(x) with its Taylor series', fontsize=16)
ax1.legend(loc='best', fontsize=14)
ax1.grid(True)
```
